```
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt

import yahoofinance as yf
import numpy as np
```

```
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```
BACK_SEARCH_DAYS = 365
endTime = dt.datetime.now();
startTime = endTime - dt.timedelta(days=BACK_SEARCH_DAYS)


```

```
pytrend.build_payload(kw_list=['Pfizer'])
# Interest by Region
df = pytrend.interest_over_time()

historical = yf.HistoricalPrices('PFE', startTime.strftime("%Y-%m-%d"), endTime.strftime("%Y-%m-%d"))
dfs = historical.to_dfs()
df2 = dfs['Historical Prices']
df2.index = pd.to_datetime(df2.index)


fig, ax1 = plt.subplots(figsize=(15,10))


color = 'tab:red'
ax1.set_xlabel('day')
ax1.set_ylabel('Google Trend', color='blue')
ax1.plot(df['Pfizer'][startTime:endTime], color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Stock', color='red')  # we already handled the x-label with ax1
ax2.plot(df2[['Close']], color='red')
ax2.tick_params(axis='y', labelcolor='red')

#fig.tight_layout()  # otherwise the right y-label is slightly clipped
#fig.

plt.show()
```
